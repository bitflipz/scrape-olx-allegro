# import & get page

```python
import requests
from bs4 import BeautifulSoup as bs
```

#### getting source from https://www.olx.pl/oferty/q-nintendo-switch/

```python
r=requests.get("https://www.olx.pl/oferty/q-nintendo-switch/")
c=r.content
```

```python
soup=bs(r.content)
```

```python
print(soup.prettify())
```

```python
last_page = int(soup.select("a[data-cy='page-link-last']")[-1].get_text().replace('\n',''))
```

# css way

```python
count = 0
for item, price in zip(soup.select("tr[class='wrap'] td[class='title-cell'] a"), soup.select("tr[class='wrap'] p[class='price']")):
    title = item.get_text().replace('\n','')
    link = item.get('href')
    try:
        price = price.get_text().replace('zł', '').replace('\n', '').replace(' ', '')
        int(price)
    except:
        price = 0
    count = count + 1
    print(f'[{count}]\ntitle: {title}\nprice: {price}')
    print(link + '\n')
```
